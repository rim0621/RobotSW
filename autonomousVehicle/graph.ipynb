```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f1=open("./pidData.txt",'r')
f2=open("./Last_data.txt","r")
```

```python
Pdata=[]
Idata=[]
Ddata=[]
PID=[]
while True:
    line=f1.readline()
    if not line : break
    tmp=line.split(",")
    Pdata.append(float(tmp[0]))
    Idata.append(float(tmp[1]))
    Ddata.append(float(tmp[2]))
    sumpid=float(tmp[0])+float(tmp[1])+float(tmp[2])
    PID.append(sumpid)
    
print(PID[0])
```

```python
i=len(Pdata)
t=[]
time=0.001
for index in range(0,i,1):
    t.append(time)
    time+=0.001
    
plt.plot(t,Pdata)
plt.title('P control')
plt.xlabel('Time 0.001')
plt.ylabel('P Data')
plt.show()

i=len(Idata)
plt.plot(t,Idata)
plt.title('I control')
plt.xlabel('Time 0.001')
plt.ylabel('I Data')
plt.show()

i=len(Ddata)
plt.plot(t,Ddata)
plt.title('D control')
plt.xlabel('Time 0.001')
plt.ylabel('D Data')
plt.show()

i=len(PID)
plt.plot(t,PID)
plt.title('PID control')
plt.xlabel('Time 0.001')
plt.ylabel('PID Data')
plt.savefig('PID1.png')
plt.show()
```

```python
Pdata=[]
Idata=[]
Ddata=[]
Errdata=[]
PID=[]
while True:
    line=f2.readline()
    if not line : break
    tmp=line.split(",")
    Pdata.append(float(tmp[0]))
    Idata.append(float(tmp[1]))
    Ddata.append(float(tmp[2]))
    Errdata.append(float(tmp[3]))
    sumpid=float(tmp[0])+float(tmp[1])+float(tmp[2])
    PID.append(sumpid)

i=len(Pdata)
t=[]
time=0.1
for index in range(0,i,1):
    t.append(time)
    time+=0.1
plt.plot(t,Pdata)
plt.title('P control')
plt.xlabel('Time 0.1')
plt.ylabel('P Data')
plt.show()

i=len(Idata)
plt.plot(t,Idata)
plt.title('I control')
plt.xlabel('Time 0.1')
plt.ylabel('I Data')
plt.show()

i=len(Ddata)
plt.plot(t,Ddata)
plt.title('D control')
plt.xlabel('Time 0.1')
plt.ylabel('D Data')
plt.show()

i=len(Errdata)
plt.plot(t,Errdata)
plt.title('err')
plt.xlabel('Time 0.1')
plt.ylabel('Err Data')
plt.show()

i=len(PID)
plt.plot(t,PID)
plt.title('PID control')
plt.xlabel('Time 0.1')
plt.ylabel('PID Data')
plt.savefig('PID2.png')
plt.show()
```
